```python
import pandas as pd

app_data = pd.read_csv('data/application_train.csv')
```

```python
%matplotlib inline
from matplotlib import pyplot as plt

target = 'TARGET'
classification_features = [col for col in app_data.columns if col not in (target,'SK_ID_CURR')]

for column in classification_features:
    print(column)
    describe_results = app_data[column].describe()
    print(describe_results)
    if (describe_results.dtype != 'object'):
        app_data[column].plot.hist()
        plt.show()
    else:
        app_data[column].value_counts().plot(kind='bar')
        print(app_data[column].unique())
    print('------')
```

```python
# handle days employed anomaly where value is 365243
anomaly_target_vals = app_data[app_data['DAYS_EMPLOYED'] == 365243][target]
non_anomaly_target_vals = app_data[app_data['DAYS_EMPLOYED'] != 365243][target]

anomaly_pct_default = 100 * anomaly_target_vals.mean()
non_anomaly_pct_default = 100 * non_anomaly_target_vals.mean()
print('Pct Default for anomalies: ', anomaly_pct_default)
print('Pct Default for non anomalies: ', non_anomaly_pct_default)
```
