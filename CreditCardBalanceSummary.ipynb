```python
import pandas as pd

credit_card_balance = pd.read_csv('data/credit_card_balance.csv')
```

```python
%matplotlib inline
from matplotlib import pyplot as plt

for col, dtype in zip(credit_card_balance.columns, credit_card_balance.dtypes):
    if dtype != 'object':
        plt.figure()
        col_data = credit_card_balance[col]
        print(col_data.describe())
        col_data.hist()
        plt.title(col)
```
